```python
# Create a dictionary where each key represents a column name and the value is a list containing the column's data
data = {
    'Product Name': ['Phone Case', 'Headphones', 'Laptop'],
    'Price': [20.99, 39.95, 799.00],
    'Quantity': [100, 50, 25]
}

# Create the DataFrame using pd.DataFrame(dictionary)
df = pd.DataFrame(data)
print(df)
```
