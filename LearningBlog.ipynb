## HOG + SVM. Detect vehicles in a video and tracking
## (Code: [HogSvmCarDetectionVideo](./HogSvmCarDetectionVideo.ipynb))

### Goal of project
1. Learn how to process videos using HOG to detect objects given a video or images
2. Take the HOG descriptors generated using training data of cars and format it to be ready for training
3. Take the data and train a SVM classifer with the descriptors
4. Create a sliding window to scan the images and grab windows that will be fed into the SVM model that was trained
5. Draw rectangles on the image where the car was predicted.
6. Combine the expected lots of rectangles into one rectangle that encapsulates the entire car
7. Figure out how to do this quickly on a video.

### Notes & learnings
* SVM stands for Support Vector Machine which is a classifer machine learning model.
* SVM groups items based on a drawing line that can be a linear or curved 'rbf kernel'. What ever is on one side of the line would classified as one class and on the other would be another class
* SVM will be trained on labeled data to allow the model to learn what class is what.
* HOG "Histogram of oriented gradients" calculates gradiates and looks for edges then provide the direction that the gradient is going. Then classifies it into a histogram of all direction vectors.
* Sobel operator which is usually a 3x3 matrix with one side of the matrix looking to multiple by -1, 0 in the center and 1 on the other side. This is done for veritcal as well. this will figure out the flow of the gradient.
* Image will be devided into cell size of 8x8 (customizable)
* Each pixel will have the calcualted gradiant magnitude and direction provided by sobel operators
* The histogram bins in hog represents the angle of the direction 
* If we picked 9 for nbins in HOG for an unsigned angle then we would get bins every 20 degrees
* Each pixel in the cell (64 in this case) will add their votes to the specific bins given there magnitude and direction of the vector
* Adding code in file to view the HOG data of a car image
* Switched to using skilearn as it provides the HOG gradient image in its API.
* Model has 99% accuracy using SVM rbf and C of 10. Could look to optimize further but is sufficient for the use case.
* Moving to implementing the sliding window that will loop through different size images throughout the image to determine if there is a car
